```python
# Create a directory for extracted files if it doesn't exist:
extracted_data_dir = "extracted_data"
os.makedirs(extracted_data_dir, exist_ok=True)

zip_file_name = "DOWNLOAD-JjUaZhE4kZ8tsbvzj31ORMejOaVHzo4aw1_3OevVftk_eq_.zip"

# Based on the uploaded output, it's likely the same name with .csv extension
csv_file_name_in_zip = zip_file_name.replace(".zip", ".csv")

with zipfile.ZipFile(zip_file_name, 'r') as z:
    # Extract the CSV file from the ZIP archive
    z.extract(csv_file_name_in_zip, path=extracted_data_dir)

chembl_raw_csv_path = os.path.join(extracted_data_dir, csv_file_name_in_zip)
df = pd.read_csv(chembl_raw_csv_path, sep=';')

# Rename columns to match expected names for filtering:
df = df.rename(columns={
    'Standard Type': 'standard_type',
    'Standard Value': 'standard_value',
    'Standard Relation': 'standard_relation',
    'Standard Units': 'standard_units',
    'Smiles': 'canonical_smiles'
})

print("--- DataFrame after column renaming (before filtering) ---")
print(df.head())
print("\n--- DataFrame describe after column renaming (before filtering) ---")
print(df.describe(include='all'))


# check for expected columns:
required_columns = ['standard_type', 'standard_value', 'standard_relation', 'standard_units', 'canonical_smiles']
missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    print(f"missing: {missing_columns}")
    print(f"missing")
    print(f"Actual columns found in the DataFrame: {df.columns.tolist()}")
    raise KeyError(f"missing: {missing_columns}")

print(f"Rows before standard_type filter: {len(df)}")
df = df[df['standard_type'].str.upper() == 'IC50']
print(f"Rows after standard_type filter: {len(df)}")
df = df[pd.to_numeric(df['standard_value'], errors='coerce').notnull()]
print(f"Rows after standard_value notnull filter: {len(df)}")
df['standard_value'] = df['standard_value'].astype(float)
print(f"Rows after standard_value type conversion: {len(df)}")
df = df[df['standard_relation'].str.strip("' ") == '=']
print(f"Rows after standard_relation filter: {len(df)}")

df = df[df['canonical_smiles'].notna()]
print(f"Rows after canonical_smiles notna filter: {len(df)}")
df = df[(df['standard_value'] > 0) & (df['standard_value'] < 1e7)]
print(f"Rows after standard_value range filter: {len(df)}")
df['pIC50'] = -np.log10(df['standard_value'] * 1e-9)
print(f"Rows after pIC50 calculation: {len(df)}")
```

```python
# Keeps only relevant columns & removes duplicates

clean_df = df[['Molecule ChEMBL ID', 'canonical_smiles', 'standard_value', 'pIC50']]
clean_df = clean_df.rename(columns={'Molecule ChEMBL ID':'ligand_id','canonical_smiles':'smiles','standard_value':'IC50_nM'})
clean_df = clean_df.drop_duplicates(subset='smiles')

clean_df.to_csv("chembl_cleaned.csv", index=False)
print("Cleaned ChEMBL CSV saved as chembl_cleaned.csv. Rows:", len(clean_df))
```

```python
descriptor_values = {}
for idx, row in clean_df.iterrows():
    mol = Chem.MolFromSmiles(row['smiles'])
    if mol is None:
        continue
    descriptor_values[row['ligand_id']] = {
        "MolWt": Descriptors.MolWt(mol),
        "LogP": Descriptors.MolLogP(mol),
        "TPSA": Descriptors.TPSA(mol),
        "RotBonds": Descriptors.NumRotatableBonds(mol),
        "HBD": Descriptors.NumHDonors(mol),
        "HBA": Descriptors.NumHAcceptors(mol)
    }

rdkit_df = pd.DataFrame(descriptor_values).T
clean_df = clean_df.set_index('ligand_id').join(rdkit_df)
clean_df.to_csv("chembl_cleaned_rdkit.csv")
print("RDKit descriptors added. Saved as chembl_cleaned_rdkit.csv")
```

```python
real_df = pd.read_csv("Controlligands - Sheet1.csv", index_col=0)
# Convert Ki to deltaG or pIC50
R = 1.987e-3
T = 298.15

epsilon = 1e-12 # Using a small number to avoid log(0)

Ki_M = (real_df['Ki_nM'] * 1e-9).replace(0, epsilon)
real_df['deltaG'] = R*T*np.log(Ki_M)
```

```python
real_desc = {}
for idx, row in real_df.iterrows():
    mol = Chem.MolFromSmiles(row['SMILES'])

    # incase smiles string is invalid.
    if mol is None:
        print(f"Warning: Could not generate molecule for ligand ID {idx} with SMILES: {row['SMILES']}")
        continue

# populate dictionary
    real_desc[idx] = {
        "MolWt": Descriptors.MolWt(mol),
        "LogP": Descriptors.MolLogP(mol),
        "TPSA": Descriptors.TPSA(mol),
        "RotBonds": Descriptors.NumRotatableBonds(mol),
        "HBD": Descriptors.NumHDonors(mol),
        "HBA": Descriptors.NumHAcceptors(mol)
    }

real_desc_df = pd.DataFrame(real_desc).T

real_df[real_desc_df.columns] = real_desc_df

# Calculate pIC50 from Ki_nM for conttol ligands
# Ensure Ki_nM is in Molar for pIC50 calculation (-log10(Molar concentration))
epsilon = 1e-12 # Define epsilon for handling zero or negative values

# Convert Ki_nM to Molar
Ki_M_for_pIC50 = real_df['Ki_nM'] * 1e-9

# Replace any values <= 0 with epsilon to avoid log(0) or log(negative)
Ki_M_for_pIC50[Ki_M_for_pIC50 <= 0] = epsilon

real_df['pIC50_from_Ki'] = -np.log10(Ki_M_for_pIC50)

print("RDKit descriptors generated and pIC50_from_Ki calculated for control ligands.")
```

```python
feature_cols = ['MolWt','LogP','TPSA','RotBonds','HBD','HBA']

if clean_df.empty:
    print("Error: clean_df is empty. Cannot train the model.")
else:
    X_train = clean_df[feature_cols].values
    y_train = clean_df['pIC50'].values

    alphas = np.logspace(-3,3,30)
    # generate sequence, adjust to your model if needed!

    # Create pipeline

    pipe = Pipeline([('scaler', StandardScaler()), ('ridge', RidgeCV(alphas=alphas, store_cv_results=True))])

    # Standardize features, select best alpha, store data

    pipe.fit(X_train, y_train)
    # Train pipeline to scale X_train using StandardScaler + fit the RidgeCV model to the scaled X_train and y_train.

    model = pipe.named_steps['ridge']

    scaler = pipe.named_steps['scaler']

    print("Ridge trained. Best alpha:", model.alpha_)
```

```python
# Calculate the correlation between Ki_nM and deltaG
correlation = real_df['Ki_nM'].corr(real_df['deltaG'])
print(f"Correlation between Ki_nM and deltaG: {correlation:.2f}")

correlation_pIC50_deltaG = real_df['pIC50_from_Ki'].corr(real_df['deltaG'])
print(f"Correlation between pIC50_from_Ki and deltaG: {correlation_pIC50_deltaG:.2f}")
```

```python
if all(col in real_df.columns for col in feature_cols):
    # check 4 required columns and redund and redundancy
    real_df_for_prediction = real_df.copy()
    real_df_valid_for_prediction = real_df_for_prediction.dropna(subset=feature_cols)
    # remove NaN rows or ones with missing values

    "checks if after dropping rows with missing feature values (which was done in the previous step with dropna()), there are any valid ligands remaining in real_df_valid_for_prediction to make predictions for. If it's empty, it means there's no data to predict, and it skips to the else block to assign NaN values."
    if not real_df_valid_for_prediction.empty:

        X_real = real_df_valid_for_prediction[feature_cols].values
        X_real_scaled = scaler.transform(X_real)

        real_df_for_prediction.loc[real_df_valid_for_prediction.index, 'predicted_penalty'] = model.predict(X_real_scaled)
    else:
        real_df_for_prediction['predicted_penalty'] = np.nan
    real_df['predicted_penalty'] = real_df_for_prediction['predicted_penalty']
    print("control ligands predicted penalties:")
    print(real_df[['Ki_nM','deltaG','predicted_penalty']])
else:
    print("Warning: real_df is missing some feature columns (e.g., MolWt, LogP, etc.) for prediction. Skipping prediction for real ligands.")
    real_df['predicted_penalty'] = np.nan # so predicted_penalty column exists even if prediction is skipped
"Save results"
real_df.to_csv("control_ligands_predicted.csv")
print("ligand predictions.")
```

```python
comparison_df = real_df.dropna(subset=['pIC50_from_Ki', 'predicted_penalty'])

if not comparison_df.empty:
    experimental_pIC50 = comparison_df['pIC50_from_Ki']
    predicted_pIC50 = comparison_df['predicted_penalty']

    # Calculate R-squared
    r_squared = r2_score(experimental_pIC50, predicted_pIC50)
    print(f"R-squared of model predictions on control ligands: {r_squared:.2f}")

    # calculate MAE and RMSE)"
    mae = mean_absolute_error(experimental_pIC50, predicted_pIC50)
    rmse = np.sqrt(mean_squared_error(experimental_pIC50, predicted_pIC50))

    "print the calculated MAE and RMSE values,"
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
else:
    print("Cannot calculate R-squared.")
```

```python
# plot results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pIC50_from_Ki', y='predicted_penalty', data=real_df, hue='predicted_penalty', size='predicted_penalty', sizes=(50, 400), palette='viridis', legend='full')
plt.title('Predicted pIC50 vs. Experimental pIC50 for control Ligands')
plt.xlabel('Experimental pIC50 (from Ki_nM)')
plt.ylabel('Predicted pIC50')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axline((0, 0), slope=1, color='red', linestyle='--', label='Ideal Prediction')
plt.legend()
plt.show()
```
